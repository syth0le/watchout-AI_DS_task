```python
from collections import Counter
import pandas as pd
from nltk.util import ngrams


read_file = 'words_count.txt'
test = 'words_count_1000.txt'
n_dict = dict()
i = 0
```

#### Отметьте разбиение N-gramm на количество слов:

```python
# укажите количество:
n = 2 
```

creating preprocessing functions for getting n-grams correctly.

```python
def preprocess(string):
    return string.split()

# можно по всем n-граммам от 2 (включительно)


def extract_n_grams(sequence, n):
    ngram = ngrams(sequence, n)
    for item in ngram:
        n_dict[item] = n_dict.get(item, 0) + 1
```

reading needful data-fle and use our functions created upper.

```python
with open(read_file) as corpus:
    for line in corpus:
        sequence = preprocess(line)
        extract_n_grams(sequence, n)
```

```python
# counting n-grams using dict
rez = Counter(n_dict)
df = pd.DataFrame(rez.most_common(5000))
df.to_csv("x{}".format(n)+'_grams.txt')
```

```python
dr = pd.read_csv("x{}".format(n)+'_grams.txt')
```

```python
dr.head(15)
```

```python
from nltk.corpus import stopwords
rus_stopwords = stopwords.words("russian")
```

Processing raw data and columns ( i will do it step by step)

```python
dr['new'] = dr['0'].apply(lambda x: str(x).strip("(')").replace("', '", ' '))
```

```python
dr
```

```python
dr['new2'] = dr['new'].apply(lambda x: [_ for _ in x.split(' ') if _ not in rus_stopwords])
```

change useless data strings to 0. It helps us to preprocess data-columns correcnly in the future movements

```python
dr['new2'] = dr['new2'].apply(lambda x: x if len(x) > n - 1 else 0)
```

```python
dr
```

```python
dr = dr.loc[dr['new2'] != 0]
```

```python
dr = dr.drop(columns=['new2', '0'])
```
