```python
morph = pymorphy2.MorphAnalyzer()
i = 0

rus_stopwords = stopwords.words("russian")
rus_stopwords.append('')
rus_stopwords.append('по')
rus_stopwords.append('с')

test_1000 = 'C:\coding\watchout!AI_DS_task\черновики'
lenta = 'lenta-ru-news.csv'
data_file = 'temptig.txt'
counting = 'words_count.txt'
d = 'words_count.txt'
test = 'copy_final.txt'
```

```python
n = time.time()
reg = re.compile('[^а-яА-Яa-zA-Z]')

f = codecs.open(test_1000, encoding='utf-8', mode='r')
for line in f.readlines():
    list_of_lists = line.replace(',', ' ').replace('-', ' ').split(' ')
    with open(counting, 'a') as w:
        for word in list_of_lists:
            if word not in rus_stopwords:
                w.write(morph.parse(reg.sub('', word))[0].normal_form + ' ')
        w.write('\n')
    i += 1
    print(i)

m = time.time()
f.close()
```

```python
print("\ntime of program working: {} sec".format(m - n))
```

```python
words = (word for line in fileinput.input(test, 'windows-1251')
         for word in re.findall(r'\w+', line.casefold()))

with open('count.txt', 'w') as w:
    for word, count in Counter(words).most_common():
        w.write("{} {},".format(word, count))
```
